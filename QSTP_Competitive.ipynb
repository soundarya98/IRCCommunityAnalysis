```python
'''New Adjacency List for whatsapp chats'''
names=[]
f=open(r"WhatsApp Chat with QSTP-Competitive Prog..txt", encoding='utf8')
contents=f.read()
contents2=contents.splitlines()
for line in contents2:
    s=line[18:]
    if(s.find(':')!=-1):
        names.append(s[:s.find(':')])
```

```python
names
```

```python
adj={}
for i in range(2, len(names)-2):
    if names[i] not in adj:
        adj[names[i]]={}
    if (names[i-2] not in adj[names[i]]) and names[i-2]!=names[i]:
        adj[names[i]][names[i-2]]=0
    if names[i-2]!=names[i]:
        adj[names[i]][names[i-2]]+=1
    if (names[i-1] not in adj[names[i]]) and names[i-1]!=names[i]:
        adj[names[i]][names[i-1]]=0
    if names[i-1]!=names[i]:
        adj[names[i]][names[i-1]]+=1
    if (names[i+1] not in adj[names[i]]) and names[i+1]!=names[i]:
        adj[names[i]][names[i+1]]=0
    if names[i+1]!=names[i]:
        adj[names[i]][names[i+1]]+=1
    if (names[i+2] not in adj[names[i]]) and names[i+2]!=names[i]:
        adj[names[i]][names[i+2]]=0
    if names[i+2]!=names[i]:
        adj[names[i]][names[i+2]]+=1   
```

```python
adj
```

```python
edges=[]
import networkx as nx
G=nx.Graph()
for name1 in adj:
    for name2 in adj[name1]:
        if name1 not in G:
            G.add_node(name1)
        if name2 not in G:
            G.add_node(name2)
        ansrow=[name1, name2, adj[name1][name2]]
        edges.append(ansrow)
        G.add_edge(name1, name2, weight=adj[name1][name2])
```

```python
nx.draw(G)
```

```python
bc=[]
bc=nx.betweenness_centrality(G)
```
