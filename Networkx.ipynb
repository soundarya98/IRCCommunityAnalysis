```python
'''Starts from here'''

root_path_orig=r"/Users/soundaryakrishnan/Documents/4-1/SOP_4-1/Scraped_Data/2018"
import os

months=os.listdir(root_path_orig)
#print(months)
months.remove('.DS_Store')

adj={}
for month in months:
    root_path=root_path_orig+"/"+month+"/"
    dirs=os.listdir(root_path)
    
    if '.DS_Store' in dirs:
        dirs.remove('.DS_Store')

    adj[month]={}
    for dir in dirs:
#         print("dir:", dir)
        files=os.listdir(root_path+ "/" +dir)
        

        for file in files:
            f= open(root_path+"/"+dir+"/"+file, encoding = "ISO-8859-1")
            contents=f.read()
            contents2=contents.splitlines()
            
            if(file not in adj[month]):
                adj[month][file]={}
                
            for i in range(len(contents2)):

                if(len(contents2[i])>=10):

                    if(contents2[i][0]=='[' and contents2[i][8]=='<'):

                        s=contents2[i]
                        name1=s[9:s.find('>')]
                        s=s[:s.find(':')]+ '~'+s[s.find(':')+1:]
                        if(s.find(':')!=-1):
                            name2=s[s.find('>')+1: s.find(':')]

                            if(len(name2)<=20 and len(name2)>=2):
                                if(name1 not in adj[month][file]):
                                    adj[month][file][name1]={}
                                if(name2 not in adj[month][file][name1]):
                                    adj[month][file][name1][name2]=[]
                                    adj[month][file][name1][name2]=0
                                adj[month][file][name1][name2]+=1
```

```python
# G=nx.Graph()
G={}
edges={}
for month in months:
    G[month]={}
    edges[month]={}
    for file in adj[month]:
        edges[month][file]=[]
        G[month][file]=nx.Graph()
        for name1 in adj[month][file]:
                for name2 in adj[month][file][name1]:
                    if(adj[month][file][name1][name2]>1):
                        #print(file, name1, name2, adj[file][name1][name2])
                        if name1 not in G[month][file]:
                            G[month][file].add_node(name1)
                        if name2 not in G[month][file]:
                            G[month][file].add_node(name2)
                        ansrow=[name1, name2, adj[month][file][name1][name2]]
                        edges[month][file].append(ansrow)
                        G[month][file].add_edge(name1, name2, weight=adj[month][file][name1][name2])
```

```python
print("starting")
bc={}
for month in months:
    bc[month]={}
    for file in G[month]:
        bc[month][file]=nx.betweenness_centrality(G[month][file])
```

```python
import operator
sorted_bc={}
for month in months:
    sorted_bc[month]={}
    for file in G[month]:
        sorted_bc[month][file] = sorted(bc[month][file].items(), key=operator.itemgetter(1), reverse=True)
```
